```python
from regraph import NXGraph, Neo4jGraph, Rule, plot_rule

graph = Neo4jGraph(
    uri="bolt://localhost:7687",
    user="",
    password="",
    node_label="Vertex",
    edge_label="LEG_TO")

# The query is slightly more complex because it needs to work on Memgraph
query = """
MATCH (j)-[:LEG_TO]->(p)-[:LEG_TO]->(i)
WHERE toInteger(p.phase % (2 * pi())) = 0
WITH p, j, i
OPTIONAL MATCH (j2)-[:LEG_TO]->(p)
WITH p, j, i, count(j2) AS in_deg
OPTIONAL MATCH (p)-[:LEG_TO]->(i2)
WITH p, j, i, in_deg, count(i2) AS out_deg
WHERE in_deg = 1 AND out_deg = 1
CREATE (j)-[:LEG_TO]->(i)
DETACH DELETE p
"""

graph._execute(query)
```
