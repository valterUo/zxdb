```python
import json
import sys; sys.path.insert(0,'..')
import os
import pyzx as zx

# The original json_to_graph does not parse inputs and outputs
from pyzx_utils import json_to_graph

fname = os.path.join('.','circuits','deutsch_n2.qasm')
circ = zx.Circuit.load(fname)#.to_basic_gates()
g = circ.to_graph()
zx.draw(g, labels=True)
```

```python
# draw the circuit with qiskit
import qiskit
from qiskit import QuantumCircuit
from qiskit.visualization import circuit_drawer

qiskit_circ = QuantumCircuit.from_qasm_file(fname)
qiskit_circ = qiskit_circ.decompose()
circuit_drawer(qiskit_circ, output='mpl', filename='deutsch_n2.png')
```
