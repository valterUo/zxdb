```python
import pyzx as zx
from pyzx.circuit.graphparser import circuit_to_graph
from zxdb.pyzx_utils import pyzx_to_networkx_tensor_graph, graph_to_quimb_tn
qubits = 2
circuit = zx.generate.CNOT_HAD_PHASE_circuit(qubits, depth = qubits)
circuit = circuit_to_graph(circuit, compress_rows=False)

nx_tn = pyzx_to_networkx_tensor_graph(circuit)
quimb_tn = graph_to_quimb_tn(nx_tn)
```

```python
quimb_tn.draw()
```

```python
print(quimb_tn.outer_inds())
out_inds = sorted(quimb_tn.outer_inds())
quimb_tn = quimb_tn.contract(output_inds=out_inds)
print(quimb_tn.data)
```
