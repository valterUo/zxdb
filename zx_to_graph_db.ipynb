```python
import json
import sys; sys.path.insert(0,'..')
import os
import pyzx as zx

fname = os.path.join('.','circuits','100.qasm')
circ = zx.Circuit.load(fname)#.to_basic_gates()
json_g = circ.to_graph().to_json()

with open("example.json", "w") as f:
    json.dump(json.loads(json_g), f, indent = 4)
```

```python
from zxdb import ZXdb

zxdb = ZXdb()

zxdb.import_zx_graph_json_to_graphdb(
    json_file_path="example.json",
    graph_id="example_graph",
    save_metadata=True,
    initialize_empty=True,  # Clear existing data for this graph_id
    batch_size=10000  # Process in batches of a million gates for better performance
    )
```
